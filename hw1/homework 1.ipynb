```python
import numpy as np
```

### task 1

```python
a = np.array([1, 2, 3, 3, 1, 6, 8, 11, 10, 7]).reshape((5,2))
a
```

```python
mean_a = a.mean(axis=0)
mean_a
```

### task 2

```python
a_centered = a - mean_a
a_centered
```

### task 3

```python
a_centered_sp = np.dot(a_centered[:,0], a_centered[:,1])
a_centered_sp
```

```python
a_centered_sp / (a.shape[0] - 1)
```

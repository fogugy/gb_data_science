```python
a_len = len(a)
np_mtrces = []
for x in range(2, 13):
    if len(np_mtrces) == 5:
        break
    if a_len % x == 0:
        np_mtrces.append(a.reshape([x, a_len // x]))
        
for i, x in enumerate(np_mtrces):
    print(i, ':')
    print(x)
```

#### Задание 2

```python
t = np.linspace(0, 10, 51)
f = np.array(list(map(lambda x: np.cos(x), t)))
```

```python
plt.plot(t, f, 'g')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.xlim([0.5, 9.5])
plt.ylim([-2.5, 2.5])
plt.show()
```

#### Задание 3

```python
x = np.linspace(-3, 3, 51)
```

```python
y1 = np.array(list(map(lambda x: x**2, x)))
y2 = np.array(list(map(lambda x: 2*x + 0.5, x)))
y3 = np.array(list(map(lambda x: -3*x - 1.5, x)))
y4 = np.array(list(map(lambda x: np.sin(x), x)))
```

```python
fig, axes = plt.subplots(2, 2)
ax1, ax2, ax3, ax4 = axes.flatten()
ax1.plot(x, y1)
ax1.set_title("График 1")
ax1.set_xlim(-5,5)

ax2.plot(x, y2)
ax2.set_title("График 2")

ax3.plot(x, y3)
ax3.set_title("График 3")

ax4.plot(x, y4)
ax4.set_title("График 4")

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)

plt.show()
```
